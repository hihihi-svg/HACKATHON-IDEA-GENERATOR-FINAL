```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Hackathon Idea Generator ‚Äî Your AI Partner for Winning Projects (Responses) - Form responses 1.csv")
df.head()
```

```python
df.columns = df.columns.str.strip()        # Remove hidden spaces
df.columns = df.columns.str.replace('\n','', regex=True)   # Remove hidden newlines
df.columns = df.columns.str.replace('\r','', regex=True)   # Remove hidden returns
df.columns = df.columns.str.replace(' +', ' ', regex=True) # Remove double spaces
```

```python
df.drop("Timestamp", axis=1)
```

```python
col = "3. On a scale of 1-5, how challenging is the 'idea generation' phase in a hackathon for you?"

counts = df[col].value_counts().sort_index()  # ensures 1,2,3,4,5 order

plt.bar(counts.index.astype(int), counts.values)
plt.xlabel("Rating (1‚Äì5)")
plt.ylabel("Number of Responses")
plt.xticks(
    [1, 2, 3, 4, 5],
    [
        "1 (Not challenging)",
        "2",
        "3",
        "4",
        "                                         5 (Extremely challenging)"
    ]
)
```

```python
col = "4. If there was an AI that gives you unique, research-based hackathon ideas instantly ‚Äî would you use it?"

counts = df[col].value_counts().sort_index()  # ensures 1,2,3,4,5 order

plt.bar(counts.index.astype(int), counts.values)
plt.xlabel("Rating (1‚Äì5)")
plt.ylabel("Number of Responses")
plt.xticks(
    [1, 2, 3, 4, 5],
    [
        "1 (Definetly)",
        "2",
        "3",
        "4",
        "         5 (No)"
    ]
)
```
